```python
import anywidget
import traitlets
from shapely import LineString

class LineStringWidget(anywidget.AnyWidget):
    _esm = "index.js"
    _css = "index.css"
    value = traitlets.List().tag(sync=True)

line = LineString([(0, 0), (1, 1), (2, 0), (3, 2)])
LineStringWidget(value=list(line.coords))
```
